```python
"""
Path Parameters & Query Parameters 

almost most of apis send and receive data through these two types of 
parameters.

"""

"""
what is parameters ?? 

=> parameters are values we pass to our API to get customized response.

Types of parameters ??

there are two types of parameters 

1) path   : /info/{name}            name which is part of URL path 
2) Query  : /search?name=Anjali     name=Anjali passed after ? 

"""


"""
1) path parameter ::: path parameter are part of the URL itself. 

they are used when your endpoint represents a specific resource -> like     
        userid , productid etc... 

"""
```

```python
"""
Add one more example 

Let's add one more dynamic route with a parameter 
"""

@app.get("/contact/{name}/")
def contact(name:str):
    return {"message":f"Hello {name}"}

"""
for testing ::: 
http://127.0.0.1:8000/greet/Anjali
"""
```

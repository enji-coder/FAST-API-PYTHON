```python
"""
What is FastApi ? 
----------------------
FastApi is a python web framework that let's you - build APIs(Application Programing Interface)
    quickly and efficiently. 
"""
```

```python
------------------------------------------------
Why we use FastApi ? 
------------------------------------------------
-> FastApi is a Much Faster (built on modern async tech)
-> More powerfull features (Like Automatic data validation)
-> Smart (Auto-generates docs and handles type checking)
-> Perfect for AI and ML Projects 
        (Used by OpenAi,Microsoft,Netflix etc...)
```

```python
------------------------------------------------
What is API ?
------------------------------------------------
An Api which is stands for Application Programing Interface 

which is connect two programs. 

e.g. Mobile Application gets weather Info it calls an API. 

An API endpoint is just a URL that return data 
    insted of a webpages.

e.g.  https://myapp.com/api/users/
```

```python
=======================================
How, FastApi works 

@app.get("/hello")
def sayHello():
    return {"message" : "welcome to FastApi"}

@app.get("/hello") :: means when someone visits 
            /hello run this function. 
```

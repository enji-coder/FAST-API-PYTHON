```python
""" 
The Core Difference — “Blocking” vs “Non-Blocking”
===============================================================
def => Synchronous (Blocking)

Code runs step by step.

Each request waits until the previous one finishes.

Suitable when you do CPU-heavy tasks or quick operations.

===============================================================
e.g.

@app.get("/normal")
def get_data():
    data = slow_database_query()  # blocks until done
    return {"data": data}

note :: If 10 users call this route at once, they wait in a queue — each handled one by one.

===============================================================
===============================================================
async def => Asynchronous (Non-Blocking)

Code runs concurrently.

Server can handle multiple requests at once while waiting for slow tasks.

Best for I/O-bound operations (like API calls, database queries, or file I/O).

e.g.. 

import asyncio

@app.get("/async_example")
async def get_data():
    await asyncio.sleep(3)  # simulates slow I/O task
    return {"msg": "done after 3 sec"}


"""
```

```python
""" 
REAL WORLD EXAMPLE
"""
import asyncio
from fastapi import FastAPI 
import time 

app = FastAPI() 
@app.get("/")
def welcome():
    return {"message":"hello welcome"}

#blocking 
# sync request waits 5 sec before starting
@app.get("/sync")
def sync_task():
    time.sleep(5)
    return {"message" : "Sync Done"}

#both start together, and both complete in ~5 sec total
@app.get("/async")
async def async_task():
    await asyncio.sleep(5)
    return {"message":"async task done"}
```

```python
""" 
The Real Difference Appears When You Run Multiple Requests at Once

Let’s simulate two users calling the endpoints simultaneously.

==> Case 1 — /sync

If 2 people call /sync together:

Request 1 starts → time.sleep(5) blocks the event loop

Request 2 waits until request 1 finishes
---> Total = ~10 seconds for both responses.

----------------------------------------------------------------

==> Case 2 — /async

If 2 people call /async together:

Request 1 starts → await asyncio.sleep(5) does not block

Request 2 starts immediately

--> Both “sleep” in parallel
--> Total = ~5 seconds for both responses.
"""
```

```python
""" 
Quick Demo Using curl

Try this in two terminals:

Terminal 1:
curl http://127.0.0.1:8000/sync

Immediately open another terminal and run:

curl http://127.0.0.1:8000/sync


You’ll notice:

The second one starts only after the first one finishes (~10 seconds total)

Now test /async:

curl http://127.0.0.1:8000/async


in both terminals quickly.
You’ll see both finish around the same time (~5 seconds).

That’s the real async power — concurrent execution.
"""
```
